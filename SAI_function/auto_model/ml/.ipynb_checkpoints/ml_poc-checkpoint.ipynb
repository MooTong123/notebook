# 2 传入训练参数
* 算法模型：str格式，二分类（binary）、多分类（multiclass）、回归（regression）
* 数据库相关信息：dict格式，包含样本数据集的相关信息（数据库地址、用户名、密码等）、输出结果的数据库相关信息（生成的评估指标图和表，生成的模型文件）
* 特征类型（feature_type）：dict格式，必有label，
    * label：str格式，列名
    * feature：dict格式，举例
        * 'feature1'：'int'(整数型)
        * 'feature2'：'float'(字符型)
        * 'feature3'：'object'(字符型)
        * 'feature4'：'datetime'(日期型)
        * 'feature5'：'category'(类别型)
        * 'feature6'：'infer'(自动推测)
        * ......
    * ignored_columns:list格式，列名，['feature7','feature8']
        
* 数据范围（data_range）:全量数据（None）或int值
* 数据划分（data_split）:数据划分比例，list格式，默认：[0.8,0.2]
* 特征处理（feature_process）:bool格式，True、False
* 时间限制（time_limit）:int格式或None，以秒为单位
* 评价指标（eval_metric）：str格式，三个机器学习任务的指标不同：
    * 二分类：accuracy（默认），precision，recall，f1，roc_auc
    * 多分类：accuracy（默认），f1_macro，f1_micro
    * 回归：root_mean_squared_error（默认），mean_squared_error，mean_absolute_error，r2
* 训练模式（presets）：str格式，基本模式（medium_quality），进阶模式（good_quality）：默认，高级模式（high_quality），极限模式（best_quality）
* 其他参数：比如项目ID，任务ID等等

```python
def get_params():
    data_range=10000
    data_split=[0.8,0.2]
    feature_process=True
    time_limit=None
    eval_metric='accuracy'
    presets='good_quality'

get_params()
```
