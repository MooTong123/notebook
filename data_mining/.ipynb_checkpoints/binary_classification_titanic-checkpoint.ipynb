# Titanic: Machine Learning from Disaster

泰坦尼克沉船生还预测任务是一个经典的机器学习任务，我们可以通过这一份用户文档，可以快速熟悉机器学习的基本流程，基本流程概括如下：

1. 定义问题(Defining the problem statement)
2. 收集数据(Collecting the data)
3. 数据探查(Exploratory data analysis)
4. 特征工程(Feature engineering)
5. 构建模型(Modelling)

# 1. 定义任务

想要做一个机器学习任务，首先要做的是了解任务是什么，我们要解决一个什么样的问题？

了解这个问题有关的背景知识可以帮助我们快速理解这个任务：

* 在1912年4月15日，泰坦尼克号在与冰山相撞后沉没，2224名乘客和船员中1502人死亡，存活率为32%。
* 这次海难造成如此多人丧生的原因之一是没有足够的救生艇供乘客和船员使用。
* 虽然在沉船中幸存下来有一定的运气因素，但一些人比其他人更有可能幸存下来。

所以任务就是：
* 运用乘客的数据构建一个预测模型来回答这样一个问题：“什么人更有可能在这次海难中存活下来？”
* 输入就是乘客的特征数据，包含乘客姓名(name)，年龄(age)，性别(gender)等等
* 输出分为两类，存活（1）和 死亡（0），所以这是一个二分类问题。




接下来我们导入我们需要的算法包，主要分为三类：
* 数据处理相关的
* 数据可视化相关的
* 机器学习建模相关（用到的时候再写）

## 2.2 理解字段信息

我们现在大致了解了训练集和测试集的样子，接下来我们需要理解每一个字段的含义,然后就可以对字段进行猜想，猜想该字段是否会影响乘客的生还率，猜想可能有对有错，都没有关系，我们可以在第3节数据探查中得到验证：

* PassengerId：乘客id(乘客id是没有实际意义的，是一个主键信息，所以应该对生存没有影响)


* Survived：是否生还(这个是label数据，必须要)
    * 1：是
    * 0：否
    
    
* Pclass：舱位（这个暗示了乘客可能是权贵，猜想贵族更有可能乘坐救生舱，对生存可能有影响）
    * 1：头等舱
    * 2：二等舱
    * 3：三等舱
    
    
* Name：姓名（名字信息可能会有含义，比如Mrs，代表了女性，对生还率有影响）


* Sex：性别（猜想会让女性先上救生舱，所以会对生还率有影响）
    * male: 男性
    * female: 女性
    
    
* Age：年龄（猜想会让老人和小孩先上救生舱，对生还率有影响）


* SibSp：在船上的兄弟/配偶数（猜想可能为了救伴侣或兄弟姐妹而耽误救援，对生还率有影响）


* Parch：在船上父母/孩子数（猜想可能为了救孩子和父母而耽误救援，对生还率有影响）


* Ticket：票号（猜想票号可能会包含舱位信息，比如头等舱等等，可能对生还率有影响）


* Fare：票价（猜想票价可能与舱位信息有关，与Pclass字段有关，可能对生还率有影响）


* Cabin：客舱号（猜想客舱号是不是代表着船上的位置，可能某些客舱离出口比较近？会先上救生舱？可能对生还率有影响）


* Embarked：登船地点（猜想可能登船地点暗示着人的地位不同？可能贵族在某一地点上传，普通人在另一地点上传？可能会对生还率有影响）
    * C: Cherbourg
    * Q: Queenstown
    * S: Southampton


# 3. 数据探查

对于传统的机器学习任务而言，高质量的数据特征能极大地提高模型预测的准确度，所以特征工程往往是传统机器学习任务的重中之重。而数据探查可以帮助建模人员快速了解数据内容、数据质量、数据分布以及数据特征间的关系，从而可以构建高质量的数据特征，完成高质量的机器学习预测任务。

数据探查主要的任务可以概括为以下方面：

1. 主要的字段（特征）分类
    * 类别型Categorical（比如性别分为男女）-- 进行编码（0和1）
    * 连续型Continous（比如年龄，是数值）-- 一般不需要处理，也可以进行分箱等处理
    * 字符型String（比如名字）-- 根据具体需要提取特征，转成数字型特征
2. 缺失值的处理
    * 删除
    * 填充
3. 探究特征间的关系
    * 主要探究的是各个特征和标记（label）数据的关系
    
总的来说，数据探查主要的工作就是对数据进行清洗，对数据进行描述（描述统计量，图表），查看数据的分布，比较数据之间的关系，培养对数据的直觉，对数据进行总结等。

* 需要根据业务逻辑的理解来判断哪些字段是类别型数据或是连续型数据，字符型字段也有可能是类别型字段，比如Sex字段。

## 3.2 缺失值的处理

在处理大规模数据时，缺失数据是经常发生的。

1. 查看缺失值的方式一般分为两种：info()或isnull()


2. 处理缺失值的方式一般分为两种：删除或替换。
    * 删除缺失值：dropna()
    * 替换缺失值：fillna()

注意：详细的缺失值处理可以在pandas文档中查看

### 3.2.1 查看缺失值

```python
# 通过info函数查看训练集的缺失值
train.info()
```

* 我们可以看到总共891行数据，其中Age、Cabin和Embarked三个字段没有达到891。同时我们也可以通过isnull()函数进行查看。

```python
# 查看训练集缺失值
train.isnull().sum()
```

```python
# 查看测试纸缺失值
test.isnull().sum()
```

### 3.2.2 处理缺失值

我们可以从上面的结果得到，训练集和测试集有缺失值的字段是：
* Age(年龄)：根据我们的推测，年龄很有可能会影响到生还率，所以删除这些缺失值是不合理的，那么我们就要进行替换填充，常用的替换策略有多种：
    * 简单策略：直接替换成该字段的中位数，平均数，众数等等，但效果可能不太理想。
    * 较复杂策略：我们可以通过业务逻辑的整理，找到该字段和其他字段的一些联系，进行选择填充。比如，我们发现Age(年龄)可能与Sex(性别)和Pclass(舱位)有关联。假如Age(年龄)字段的平均值为30，但我们发现头等舱的女性的平均年龄是20，那么当一个样本符合头等舱和女性两个条件时，预测她的年龄为20岁就比预测她为30岁更加准确。


* Cabin(客舱号):根据我们的猜测，客舱号可能与逃生出口的远近有关，但是因为没有这方面的相关资料，并且缺失率达到了77%（891个样本中687个缺失值），所以我们选择直接删除这个字段。


* Embarked(登船地点)和Fare（票价）：这两个字段缺失值都很小，所以我们暂时用众数来替代.

#### 3.2.2.1 处理Age(年龄)字段

我们经过业务逻辑的整理，猜测到Age字段和Sex(0、1)与Pclass(1、2、3)有关，所以有6种组合关系，下面通过代码进行实现：

```python
# 首先查看训练集前10行数据
train.head(10)
```

* 我们观察到第5行的Age列有一个缺失值NaN。

```python
# 通过groupby分组聚合求出Age的中位数
train["Age"].groupby([train["Sex"], train["Pclass"]]).median()
```

* 所以上面观察到的第5行数据是Pclass=3，Sex=male，所以Age应该填充为25

```python
# 使用fillna函数进行训练集和测试集的Age字段的填充
train["Age"].fillna(train["Age"].groupby(
    [train["Sex"], train["Pclass"]]).transform("median"), inplace=True)
test["Age"].fillna(test["Age"].groupby(
    [test["Sex"], test["Pclass"]]).transform("median"), inplace=True)
```

# 4. 特征工程

我们为什么要做特征工程呢？
* 选对模型有用的特征，让模型达到尽可能大的性能（准确率）
* 剔除掉无用特征，减少模型的过拟合，增加模型的泛化能力
* 对有用的特征做相应的变化处理，减少模型复杂性，增加模型的性能

常见的特征工程有什么步骤呢？
* **特征采集评估**：主要是评估特征的覆盖率、特征的完整性、特征的有效性等。


* **特征预处理**：主要是对特征进行定性分析，缺失值、异常值、基本特征编码等


* **特征无量纲化**：主要是将特征从一个范围映射到新的空间，调整数据的分布，帮助机器学习模型更好的学习特征，常见的方法为min-max标准化，z-score标准化等


* **特征选择**：选择对模型更有效的特征，常见的方法有基于统计的过滤法（方差选择，皮尔逊相关系数等）、嵌入法（一些机器学习模型比如决策树就会自动选择特征）


* **高级特征变换**：对特征进行降维、分箱、构建新的特征等高级技巧，转换出对模型更有效的特征


讨论：
* **特征采集评估**：是构建数据时要采集哪些特征需要考虑的问题，本项目中已经有了构建好的特征表了，所以不需考虑这个问题。


* **特征预处理**：这里的缺失值处理就是我们在数据探查中做的缺失值处理，基本特征编码有两种：独热编码（one-hot encoding）和 class编码。
    * **独热编码**：例如，男和女的编码分别为[（0,1）,（1,0）]，适用于有距离度量的模型（比如LR，SVM等）
    * **class编码**:例如，男和女的编码分别为[0,1],适用与对距离不敏感的模型，比如决策树模型。


* **特征无量纲化**：在一些数据集中，可能因为特征尺度差异巨大，导致模型收敛慢或无法收敛，所以做了同一纬度的纲化，可以加速模型的收敛过程。比如，一个特征的范围是[0,10000],另一个特征的范围[0,10]，这就是特征尺度的差异十分巨大，经过无量纲化，可以把他们转变成为[0,1]里面的数字，这样就可以加速模型的收敛。在本例中，没有这样的特征，所以不需要做无量纲化。


* **特征选择**：特征选择的主要目的是选择有意义的特征，减少模型复杂度。一些集成的机器学习算法中，比如决策树相关的模型（lightgbm，xgboost等）已经内置了特征选择的功能，所以我们也不需要做该方面的工作。


* **高级特征变化**：常见的方法为降维，在本例中，我们经过筛选选出了8个特征，所以不需要做降维处理，如果你的数据有几百个特征等等，那就需要做特征降维处理。我们还可以根据业务理解来构建新的特征，比如我们有两个特征（身高和体重），该如何判断一个是胖是瘦？其实我们可以根据业务理解，构建一个新的特征BMI（身体质量指数），就可以更好的衡量一个人的胖瘦。

接下来，我们使用特征预处理的class编码和高级特征变化的构建新特征，对特征进行处理。

```python
# 用正则表达式进行称谓提取
train["Title"] = train.Name.str.extract('([A-Za-z]+)\.', expand=False)
test["Title"] = test.Name.str.extract('([A-Za-z]+)\.', expand=False)
```

```python
# 将提取的称谓进行整理，归纳为常见的称谓和Rare信息
for dataset in (train, test):
    dataset["Title"] = dataset["Title"].replace(
        ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

# 根据称谓Title查看生还率
train[["Title", "Survived"]].groupby("Title", as_index=False).mean()
```

# 5. 构建模型

现在我们已经构建好了特征向量，下一步就是训练一个机器学习模型并在测试集上检测模型的效果。如何选择一个合适的机器学习模型是至关重要的，首先最关键的就是根据问题的本质来选择对应的机器学习模型，我们可以把问题粗略的分为三大类：分类、回归、聚类。

* 分类问题：一种对离散型随机变量建模或预测的监督学习算法，比如，邮件过滤、金融欺诈等案例。常见的模型包含：逻辑回归、朴素贝叶斯、支持向量机、K近邻、分类树（集成方法，比如随机森林、GBDT等）、深度学习（图像分类、文本分类）等。


* 回归问题：一种对数值型连续随机变量进行预测和建模的监督学习算法。比如，房价预测、股票走势等案例。常见的模型包含：线性回归、回归树（集成方法、随机森林、GBDT等）、深度学习等。


* 聚类问题：一种无监督学习算法，该算法基于数据的内部结构寻找观察样本的自然族群，即把相似性很高的样本分成一类。比如，新闻聚类、文章推荐等。常见的模型包含：K均值聚类、层次聚类等。

讨论：
1. 选择合适的机器学习算法，最先考虑的是要解决的问题属于分类、回归、聚类哪一类问题。比如，我们这里演示的问题（泰坦尼克号预测生还），就是一个分类问题（生或死），所以我们就要采用适合分类问题的机器学习算法。

2. 其次需要考虑的问题，便是模型的性能和其他因素（比如可解释性、复杂性、数据集大小、特征维度、训练时间和成本、推理时间等）。
    
* **(1) 性能**：模型的性能是选择模型的最基本因素，一般来说会优先选择性能最大的模型算法。需要注意的是，根据不同的问题会有不同的度量标准，选择一种良好的度量来评估模型的性能是一件重要的任务，常见的度量标准有正确率、查全率、均方误差等。

    
* **(2) 可解释性**：在很多实际情况下，客户需要对模型结果进行解释。但很多算法就像黑盒一样，无论结果如何，都很难进行解释，比如神经网络。如果对可解释性要求比较高的话，决策树相关的算法是很好的选择。


* **(3) 复杂性**：一般而言，模型复杂性越高，性能就越好，但成本也越高。并且复杂性和可解释性成反比，模型越复杂，解释结果就越困难。


* **(4) 数据集大小**：可用的训练数据集量也是选择模型时需要考虑的因素之一。比如神经网络在大量训练数据中，有很好的性能；但是如果你的训练集只有很少一部分，比如几百条，那么无论你在神经网络怎么调优，都不会有很好的效果，反而采用一些简单的模型，比如逻辑回归，可能就有很好的效果。


* **(5) 特征维度**：特征维度也是选择模型时需要考虑的因素，一些算法不适合高纬度的数据，比如K近邻。如果特征维度过高，会导致模型过于复杂，所以也可考虑特征降维等方法来减小特征维度。
    
  
* **(6) 训练时间和成本**：训练一个的模型是需要成本的。比如，训练一个准确率达到95%的模型，需要10万；训练一个准确率达到93%的模型，需要1万，那么你会选择哪种模型呢？答案取决于你的业务。比如，你做的是一个金融欺诈模型，召回率越高越好，如果遗漏任何一个负样本，可能都会造成很大的经济损失，所以不会考虑成本的问题，性能越高越好；但是一些其他的模型，比如推荐系统，95%的性能和93%的性能，对于用户来说没有很大的差距，但对于公司而言，成本是翻了好几倍的，所以选择一个成本和性能相平衡的解决方案是很重要的。
  
  
* **(7) 推理时间**：训练好一个模型并做出预测需要多长时间？比如，一个自动驾驶系统，需要做出实时决策，如果预测时间太长，将会导致无法预料的后果，所以必然不会选择模型复杂度大的模型。如果模型太大，也可以选择**模型蒸馏**等方式使模型变得简单。

## 5.2 模型训练

基本的模型构建过程如下：
* 实例化模型学习器
* 拟合训练集

## 5.5 模型调优

我们以LightGBM为例，来演示如何进行模型调优。

LightGBM一些重要的模型参数，如下所示，其中最重要的是4个参数（learning_rate, n_estimators, max_depth, num_leaves），其他的参数是控制过拟合现象的：

* **learning_rate**:默认设置为0.1，一般设置在0.01-0.1之间。选择较小的学习率可以获得稳定较好的模型性能。


* **n_estimators**：boosting的迭代次数，默认值为100.一般根据数据集和特征维度选择100-1000之间。更保守的做法是设置一个较大的值配合early_stop来让模型根据性能自动选择最好的迭代次数。


* min_split_gain: 最小分类增益，默认是0。一般不建议调整这个参数，增大这个值会得到相对浅的树深，只有在过拟合严重时，参会考虑增大这个参数。


* min_child_samples：分支节点的最小样本数，默认值为20。一般小数据集默认的20就不需要更改了，但大数据集需要增大这个值。


* min_child_weight：分裂节点的最小权重，默认值是0.001。作用和min_child_samples类似，在大数据集时需要调整。


* **max_depth**：树的最大深度，是防止过拟合的最重要的参数，一般限制为3-5之间，对模型的性能和泛化能力有决定性作用。


* **num_leaves**：一棵树上的叶子节点数，默认设置为31。默认值为31，和max_depth配合使用，一般设置为[0,2^max_depth]的一个数值，对模型性能影响很大。


* subsample:训练样本的子采样比例，默认是1。不建议过度的精细调整，在[0.8.0.9,1]中搜索就足够了。


* colsample_bytree：训练特征的子采样比例，默认是1。和subsample一样，不建议过度的精细调整，在[0.8.0.9,1]中搜索就足够了。


* reg_alpha：L1正则化系数，默认是0。一般我们取0-1000的范围去进行调参。如果优化出来这个参数数值过大，则说明有一些不必要的特征可以剔除，可以先做特征筛选后再进行调参，然后调节出来模型效果好的时候reg_alpha是个相对小的数值，那我们对这个模型的信心会大很多。


* reg_lambda：L2正则化系数，默认值是0。般也是在0-1000的范围去进行调参。如果有非常强势的特征，可以人为加大一些reg_lambda使得整体特征效果平均一些，一般会比reg_alpha的数值略大一些，但如果这个参数大的夸张也需要再查看一遍特征是否合理。


### 5.4.1 默认参数生成的模型

```python
# 初次尝试，默认参数
lgb_clf_1 = lgb.LGBMClassifier(
    n_estimators=100, learning_rate=0.1, max_depth=-1, num_leaves=31)
lgb_clf_1.fit(train_x, train_y)
```

```python
# 查看测试集的准确率
lgb_clf_1.score(x_test, y_test)
```

以上，我们用默认参数进行了训练并检验测试集，接下来我们用网格搜索（grid search）来确定合适的 learning_rate, n_estimators, max_depth, num_leaves,并注意关系 (num_leaves <= 2^max_depth - 1)。

### 5.4.2 网格搜索进行参数调优

1. 一般而言，我们先进行 learning_rate 和 n_estimators 两个参数的调整。经验主义告诉我们，我们可以设置较大的 n_estimators 和较小的 learning_rate，再使用早停（early_stop），即可以得到一个较好的精度。


2. 然后我们再根据经验指定一个大概范围的 max_depth, 一般小数据集可以为3-5，较大数据集可以指定为5-8。max_depth 太小则会导致过拟合现象，适当增大 max_depth 可以增加模型的泛化能力。


3. 最后我们再调整根据 max_depth 和 num_leaves 的关系，来确定 num_leaves的范围。

```python
from sklearn.model_selection import GridSearchCV

# 使用网格搜索进行参数确定
param_grid = {'learning_rate': [0.01, 0.03, 0.05],
              'n_estimators': [100, 200, 300]}

# 较小数据集，指定max_depth为3
lgb_estimator = lgb.LGBMClassifier(max_depth=3)

# 使用网格搜索进行参数搜索
lgb_clf_2 = GridSearchCV(lgb_estimator, param_grid, cv=5)

# 传入验证集，设置早停early_stop
lgb_clf_2.fit(train_x, train_y, early_stopping_rounds=10,
              eval_set=(valid_x, valid_y), verbose=False)

# 显示最好的参数
lgb_clf_2.best_params_
```

调整完两个参数后，下面我们再调整 max_depth。

```python
# 使用网格搜索进行参数确定
param_grid = {'max_depth': [2, 3, 4, 5]}

# 根据上面的结果，指定 learning_rate 和 n_estimators
lgb_estimator = lgb.LGBMClassifier(learning_rate=0.01, n_estimators=200)

# 使用网格搜索进行参数搜索
lgb_clf_2 = GridSearchCV(lgb_estimator, param_grid, cv=5)

# 传入验证集，设置早停early_stop
lgb_clf_2.fit(train_x, train_y, early_stopping_rounds=10,
              eval_set=(valid_x, valid_y), verbose=False)

# 显示最好的参数
lgb_clf_2.best_params_
```

调整完3个参数后，下面我们再调整 num_leaves, 我们根据(num_leaves <= 2^max_depth - 1)，求出 num_leaves 的最大值为7。

```python
# 使用网格搜索进行参数确定
param_grid = {'num_leaves': [3, 4, 5, 6, 7]}

# 根据上面的结果，指定 learning_rate，n_estimators和 max_depth
lgb_estimator = lgb.LGBMClassifier(
    learning_rate=0.01, n_estimators=200, max_depth=3)

# 使用网格搜索进行参数搜索
lgb_clf_2 = GridSearchCV(lgb_estimator, param_grid, cv=5)

# 传入验证集，设置早停early_stop
lgb_clf_2.fit(train_x, train_y, early_stopping_rounds=10,
              eval_set=(valid_x, valid_y), verbose=False)

# 显示最好的参数
lgb_clf_2.best_params_
```

```python
# 求出目前测试集的准确率
lgb_clf_2.best_estimator_.score(x_test, y_test)
```

经过我们简单的4个参数调优演示，测试集的准确率从83%提高到了94%，提高了11%，这也充分说明了参数调优的必要性。
